```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
```

```python
# Extracting Tesla Stock Data Using yfinance
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
# Extracting Tesla Revenue Data Using Webscraping 
# Step 1: Fetch and parse the page
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
data = requests.get(url).text
soup = BeautifulSoup(data, 'html.parser')

# Step 2: Find the Tesla Quarterly Revenue table
tables = soup.find_all("table")

df_tesla_revenue = None
for table in tables:
    if "Tesla Quarterly Revenue" in table.text:
        df_tesla_revenue = pd.read_html(str(table))[0]
        break

# Step 3: Clean the table
if df_tesla_revenue is not None:
    df_tesla_revenue.columns = ["Date", "Revenue"]
    df_tesla_revenue["Revenue"] = df_tesla_revenue["Revenue"].str.replace("$", "", regex=False).str.replace(",", "", regex=False)
    df_tesla_revenue.dropna(inplace=True)
    df_tesla_revenue = df_tesla_revenue[df_tesla_revenue["Revenue"] != ""]
    display(df_tesla_revenue.head())
else:
    print("❌ Tesla revenue table not found in the provided IBM course URL.")
```

```python
# Extracting GameStop Stock Data Using yfinance
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")
gme_data.reset_index(inplace=True)
gme_data.head()
```

```python
# Extracting GameStop Revenue Data Using Webscraping
# Step 1: Corrected URL (without the ending period)
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
html_data = requests.get(url).text
soup = BeautifulSoup(html_data, "html.parser")

# Step 2: Locate and extract the table
df_gme_revenue = None
tables = soup.find_all("table")

for table in tables:
    if "GameStop Quarterly Revenue" in table.text:
        df_gme_revenue = pd.read_html(str(table))[0]
        break

# Step 3: Clean the data
if df_gme_revenue is not None:
    df_gme_revenue.columns = ["Date", "Revenue"]
    df_gme_revenue["Revenue"] = df_gme_revenue["Revenue"].str.replace("$", "", regex=False).str.replace(",", "", regex=False)
    df_gme_revenue.dropna(inplace=True)
    df_gme_revenue = df_gme_revenue[df_gme_revenue["Revenue"] != ""]
    display(df_gme_revenue.head())
else:
    print("❌ GameStop revenue table not found. Please check the page structure.")
```

```python
# Tesla Stock and Revenue Dashboard
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
plt.plot(tesla_data['Date'], tesla_data['Close'], label='Tesla Stock Price')
plt.title('Tesla Stock Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(pd.to_datetime(df_tesla_revenue['Date']), df_tesla_revenue['Revenue'].astype(float), label='Tesla Revenue', color='orange')
plt.title('Tesla Revenue')
plt.xlabel('Date')
plt.ylabel('Revenue (Millions)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#GameStop Stock and Revenue Dashboard
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
plt.plot(gme_data['Date'], gme_data['Close'], label='GameStop Stock Price')
plt.title('GameStop Stock Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(pd.to_datetime(df_gme_revenue['Date']), df_gme_revenue['Revenue'].astype(float), label='GameStop Revenue', color='green')
plt.title('GameStop Revenue')
plt.xlabel('Date')
plt.ylabel('Revenue (Millions)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
